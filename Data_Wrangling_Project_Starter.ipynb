## 3. Clean data
Clean the data to solve the 4 issues corresponding to data quality and tidiness found in the assessing step. **Make sure you include justifications for your cleaning decisions.**

```python
# FILL IN - Make copies of the datasets to ensure the raw dataframes 
# are not impacted
df_movies_clean = df_movies.copy()
df_credits_clean = df_credits.copy()
```

### **Quality Issue 1: release_date have object datatype instade of datetime**

```python
# FILL IN - Apply the cleaning strategy
df_movies_clean ['release_date'] = pd.to_datetime(df_movies['release_date'])
```

```python
# FILL IN - Validate the cleaning was successful
df_movies_clean .info()
```

Justification: I converted release_date from string to datetime datatype

### **Quality Issue 2: there is NULL values**

```python
#FILL IN - Apply the cleaning strategy
df_movies_clean .dropna(inplace=True)
```

```python
#FILL IN - Validate the cleaning was successful
df_movies_clean .info()
```

Justification: I dropped the Null value raws

### **Tidiness Issue 1: The column 'genres' and 'productions_companies' have multiple value in the same row of data.**

```python
#FILL IN - Apply the cleaning strategy
df_movies_clean ['genres'] = df_movies_clean ['genres'].apply(lambda x: x.split('|')[0])
df_movies_clean ['production_companies'] = df_movies_clean ['production_companies'].apply(lambda x: x.split('|')[0])
```

```python
#FILL IN - Validate the cleaning was successful
df_movies_clean.head(7)
```

Justification: I chose to eliminate the values after the first '|' symbol.

### **Tidiness Issue 2: after combining the datasets we will have duplicate columns "title" and " cast"**

```python
#FILL IN - Apply the cleaning strategy
df_movies_clean=df_movies_clean.drop(['original_title'], axis=1)
df_credits_clean =df_credits_clean .drop(['cast'], axis=1)
```

```python
#FILL IN - Validate the cleaning was successful
df_movies_clean.head(7)
```

```python
df_credits_clean.head(7)
```

Justification: i dropped the "title" column from the movies dataset and "cast"  column from credits dataset

### **Remove unnecessary variables and combine datasets**

Depending on the datasets, you can also peform the combination before the cleaning steps.

```python
#FILL IN - Remove unnecessary variables and combine datasets
df_movies_clean = df_movies_clean.drop(['cast','homepage','tagline','keywords','overview','imdb_id'], axis=1)
```

```python
df_movies_clean.head(3)
```

```python
df_movies_clean=df_movies_clean.drop_duplicates()
```

```python
df_credits_clean=df_credits_clean.drop_duplicates()
```

```python
combined= pd.merge(df_credits_clean ,df_movies_clean , left_on='movie_id', right_on='id' )
combined.drop('id', axis=1, inplace=True)
combined.head(5)
```
